```python
#Average Degree:
def AvDegree(G1):
    r = 0
    N = G1.number_of_nodes()
    for i in G1.nodes():
        r = r+G1.degree(i)
    A = r/N 
    return A

#Since we do not have calculated the density yet, we took the list of degree and made the average
```

```python
#Density
def Density(G1):
    N = G1.number_of_nodes()
    m = len(G1.edges)
    c = 2*m/(N*(N-1))
    return c
#So for the density, we need to find the ratio between the number of edges and the max number of edges that the graph can contain, we simply found
#the number of edges and wrote the formula with the number of nodes
```

```python
#Inverse Distance
def InvDistance(G2,i,j):
    A = nx.adjacency_matrix(G2)
    N = G2.number_of_nodes()
    inv_dist=0
    M = scipy.sparse.identity(N)
    for l in range(N):    
        M = M * A
        # M = A ** (l+1) (default code, not efficient)
        if M[i,j]==1 and inv_dist==0:
            inv_dist=1/(l+1)
    return inv_dist

print(InvDistance(G2,3,94))
# print(nx.adjacency_matrix(G2).shape)
# print(scipy.sparse.identity(181) * nx.adjacency_matrix(G2) == nx.adjacency_matrix(G2))
```

```python
def average_inverse(G2):
    N = G2.number_of_nodes()
    Temp_sum = 0
    for i in range(N):
        for j in range(i+1, N):
            Temp_sum = Temp_sum+InvDistance(G2, i, j)
    if Temp_sum == 0:
        result = 'No Edges'
    else:
        result = (N*(N-1))/(Temp_sum*2)
    return result
```

The aim of week three was to understand our graph better using distribution of centraities. We decided that Closeness centrality would have been a good choice in order to compute each characters position in the network, capturing the average distance between each vertex and every other vertex in the network. A low closeness centrality means that a person is directly connected from most others in the network. In contrast, vertices in very peripheral locations may have high closeness centrality scores. After that we calculated the most central of all.

```python
# Week 3

import functools

import networkx as nx
from networkx.exception import NetworkXError
from networkx.utils.decorators import not_implemented_for

def closeness_centrality(G2,u = None, distance = None, wf_improved= True ):
    global closeness_dict
    if G2.is_directed():
        G2 = G2.reverse()  # create a reversed graph view

    if distance is not None:
        # use Dijkstra's algorithm with specified attribute as edge weight
        path_length = functools.partial(
            nx.single_source_dijkstra_path_length, weight=distance
        )
    else:
        path_length = nx.single_source_shortest_path_length

    if u is None:
        nodes = G2.nodes
    else:
        nodes = [u]
    closeness_dict = {}
    data_centralities = []
    for n in nodes:
        sp = path_length(G2, n)
        totsp = sum(sp.values())
        len_G2 = len(G2)
        _closeness_centrality = 0.0
        if totsp > 0.0 and len_G2 > 1:
            _closeness_centrality = (len(sp) - 1.0) / totsp
            # normalize to number of nodes-1 in connected part
            if wf_improved:
                s = (len(sp) - 1.0) / (len_G2 - 1)
                _closeness_centrality *= s
        closeness_dict[n] = _closeness_centrality
    if u is not None:
        return closeness_dict[u]
    return closeness_dict 
```

We want now to investigate how much the graph is robust with respect to the deletion of nodes. We investigate the connectivity by the size of the largest component.

We will investigate two different perturbations:

error: randomized uniformly. All nodes have the same probability to fail

attacks: the attacker prefer to attack nodes that are central (notion of centrality need to be chosen)

```python
def robustness_uniform(G1):
    
    H=G1.copy()  # We want a copy to work with. Without the copy, we would change the graph and forget the original graph.
    
    N=H.number_of_nodes()
    list=[]
    
    for n in range(N-1):
        
        Nodes=H.nodes()   # find the list of remaining nodes
        i=random.sample(Nodes,1)[0]  # pick one of them 
        
        H.remove_node(i)  # take it out
        
        largest_cc=max(nx.connected_components(H), key=len) # what is the largest component

        len_largest_cc=len(largest_cc)
        
        list.append(len_largest_cc)
    
    return list
```

```python
list_robustness=robustness_uniform(G1)
plt.plot(list_robustness)
```

The previous compilation is made to be randomized. Depending on which nodes are chosen, we obtain different curves therefore it is not clear which part of the curve is due to the randomness and which part to the network.

To understand better, we do many simulations and we average on them the values to obtain a graph depending only on the network.

```python
def statistical_uniform(G,L):

    data1=robustness_uniform(G)
    
    for l in range(L):
           
        data2=[]
        local=robustness_uniform(G)
        
        for a in range(len(local)):
            data2.append(local[a]+data1[a])
        
        data1=data2.copy()
                       
    Result=[i/(L+1) for i in data1]
    
    return Result
```

```python
def list_to_proba(list):
    S=sum(list)
    
    if S==0:
        Result=[1/len(list) for i in list]
    else:
        Result=[i/S for i in list]
    
    return Result
```

For week five we implemented a page rank alghoritm  to calculate the PageRank, which is uselful to understand the importance of the Nodes that we are analyzing. Then we used the built-in function from NetworkX to compare our results. PageRank will assign a score of importance to every single node. And the assumption that it makes is that important nodes are those that have many in-links from other important nodes.

```python
# Week 5

#PageRank Code

def PageRank(alpha):

    Rp = {} #initial list in which all nodes have the same probability
    R = {} #temporary variable for each iteration of PageRank
    R_check = {} #dictionary with rounded values of PageRank
    iterations = 0
    k_out = {} #outdegree of each node
    neighbors = {} #list of neighbors of each node
    R_dict = {} #dictionary with the values of each iteration of PageRank
    check = False #is the stopping condition, it check = True the PageRank stops 
    R_output = {} 
    nodes = G2.nodes()

    constant_pr = alpha/len(nodes) #first value to compute PageRank

    #here we insert the zero probabilities in the list
    for i in nodes:
        Rp[i] = 1/len(nodes)
    R_dict[0] = Rp
    R_check[0] = Rp

    #calculate the degree of nodes and their neighbors
    for node in nodes:
        k_out[node] = nx.degree(G2, node)
        sublist = []
        for j in nx.neighbors(G2,node): 
            sublist.append(j)
        neighbors[node] = sublist
        
    while check != True:
        iterations +=1 #move to the following iteration
        R = {} #reinitialize the temporary PageRank value
        Temp_R_check = {} #reinitialize the temporary PageRank value used to check the convergence of PageRank
        for node in nodes: #calculate PageRank for each node
            pred_sum = 0
            for j in neighbors[node]:
                pred_sum += (R_dict[iterations-1][j])/(k_out[j])
            R[node] = constant_pr + (1 - alpha)*(pred_sum)
            Temp_R_check[node] = round((constant_pr + (1 - alpha)*(pred_sum)),6)
        R_check[iterations] = Temp_R_check #add the approximated values to the dictionary 
        if R_check[iterations-1] == R_check[iterations]: #verify if the actual values and the previous ones converge
            check = True #convergence stops the counter
        R_dict[iterations] = R #add the values of PageRank to the dictionary that maps the iterations of PageRank
    
    #we convert the final PageRank to a dictionary where nodes are keys and values of PageRank are values
        
    return [R_dict[iterations], iterations]

print('This is the page rank of our graph:'+ '\n')

PR = PageRank(0.15)

print(PR[0])
print('\n')

print('That converges to the iteration number: ')

print(PR[1])
print('\n')

print('Now we can compare our own algorithm with the NetworkX function:' + '\n')

print('The NX page rank is : ' + '\n')

print(nx.pagerank(G2, alpha = 0.85, tol = 1e-06))
print('\n')
    
```

# WEEK 7

The goal of week seven is to calculate the topological similarity indeces.
We created a function that calculates every index for every node pair, said function returns a DataFrame of node pairs as rows and the five topological indices as columns. All of this obviosly happens only for unconnected node pairs (hence we iterate over a list of missing links).
We then implemented a function in order to computed the 6 link likelihood scores, one for each 5 indices and one obtained as the aritmetic mean of the indices (after we rescaled each column between 0 and 1).
Finally, for each of the six scores, we identified as missing links the node pairs yelding the largest 10 values.

```python
import pandas as pd
import tqdm as tqdm
import itertools as it
```

```python
#Topological indeces

def top_index(G, ind_list: list): 
    diameter = nx.diameter(G)

    indices_dict = {}
    for u in G.nodes:
        for v in G.nodes:
            if u != v and (u, v) not in G.edges:
                indices_dict[(u, v)] = []
    missing_links = indices_dict.keys()

    if len(ind_list) == 0:
        return top_index(G, ['CN', 'JI', 'PA', 'AA', 'RA'])
   
    for ind in ind_list:
        
        if ind not in ['CN', 'JI', 'PA', 'AA', 'RA']:
            return "Error, you must give as input only the following indeces in this form: ['CN', 'JI', 'PA', 'AA', 'RA']."
        
        if ind == 'CN':
            for link in missing_links:
                CN_list = []
                for i in nx.common_neighbors(G, link[0], link[1]):    
                    CN_list.append(i)
                indices_dict[link].append(len(CN_list))
        
        if ind == 'JI':  
            for link in missing_links:
                for i in nx.jaccard_coefficient(G, [link]):
                    JI_value = i[2]
                indices_dict[link].append(JI_value)

        if ind == 'PA':
            for link in missing_links:
                for i in nx.preferential_attachment(G, [link]):
                    PA_value = i[2]
                indices_dict[link].append(PA_value)
        
        if ind == 'AA':    
            for link in missing_links:
                for i in nx.adamic_adar_index(G, [link]):
                    AA_value = i[2]
                indices_dict[link].append(AA_value)
        
        if ind == 'RA':    
            for link in missing_links:
                for i in nx.resource_allocation_index(G, [link]):
                    RA_value = i[2]
                indices_dict[link].append(RA_value)
            
    data_frame = pd.DataFrame.from_dict(indices_dict, orient='index', columns=ind_list)
    return data_frame
```

For week 9, we decided that the most appropriate random graph to use was the Preferential attatchment. (it acts in the same way as uniform attachement except for when it chooses the links. It uses weights proportional to the degree) 

```python
def create_pa(vertex_num: int,edge_num: int):
    
    m=edge_num
    l=vertex_num
    G = nx.complete_graph(m) # start with a complete graph s.t. every initial node can be reach
    
    c = (m-1)
    
    for i in range(l):
        
        if c==0: #initialization if there is only one node
            c=c+1
            G.add_edge(1,0,length=1)
        else:
            c=c+1

            liste_deg=G.degree(range(c)) # take the degree
            liste = np.random.choice(range(c), m,liste_deg) # pick in the list betzeen 0 and c-1, m element with weight p
            
            for n in liste:
                G.add_edge(c,n,length=1)    
                 
    return G
```

The aim of week ten is to simulate several dynamics of propagation in a network. (SIR in our case)

Here is the code of the SIR model.

```python
# Week 10

def simulation_SIR_network(G,I,T,beta,mu,graph):
    # initialisation of the number of each node
    s = 0
    i = 0
    r = 0
    n0 = G.number_of_nodes()
    for node in G.nodes(): # Initialization of the agent in I as infected and other agent as susceptible
        if node in I:
            G.nodes[node]['category'] ='type_I'
            i = i + 1
        else:
            G.nodes[node]['category'] ='type_S'
            s = s + 1
            
    color_map = {'type_S':'b', 'type_I':'r','type_R':'g'} # 3 colors

    if graph==True: # plot or not
        # plot with the colors
        nx.draw_circular(G,node_size=50, node_color=[color_map[G.nodes[node]['category']] for node in G]) 
        plt.show()
        
    t = 0 # create variable to keep track of the evolutions
    data_S = []
    data_I = []
    data_R = []
    n = n0 - 1
    
    # initialization
    data_I.append(i)
    data_R.append(r)
    data_S.append(s)
    while (t <= T) and (i != 0): 
        t = t+1

        for node in G.nodes(): # go over the node
            switch = 0 # we want to allow only one change for a given node by each iteration.

            # if the node is recovered
            if (G.nodes[node]['category'] == 'type_I') and (random.uniform(0, 1) <= mu):           
                G.nodes[node]['category'] = 'type_R'
                i = i-1
                r = r+1
                switch = 1 # changed
                
            # if the node is susceptible
            if (G.nodes[node]['category']=='type_S') and (switch == 0): 
                for neighbor in G.neighbors(node): # go over neighbors of the current node
                    if (G.nodes[neighbor]['category']=='type_I') and (random.uniform(0, 1) <= beta) and (switch == 0): 
                        G.nodes[node]['category']='type_I'
                        
                        #update number of nodes in each category
                        s = s-1
                        i = i+1
                        switch = 1 # changed
            
         # Plotting part
        if graph == True: # plot or not
            nx.draw_circular(G,node_size=50, node_color=[color_map[G.nodes[node]['category']] for node in G])
            plt.show()
                
        # update the list for the current stage
        data_I.append(i)
        data_R.append(r)
        data_S.append(s)
        
    # list of iteration for plotting
    data_x = range(t+1)
    
    return data_x, data_S, data_I, data_R, G # return list of time, number of S/I/R
```

Let us now investigate different scenarios of outbreak in our graph (we can do so simply by changing the number of infected nodes in the graph at the start of infection).

We will see the plot when the infection starts from a single node, when it starts from the entire cluster, when it starts from some element within a cluster or when it starts from some nodes in each connected component.

```python
T=100
beta=3/4
mu=1/10

# infection starts from single node
data_x_SIR, data_S_SIR, data_I_SIR, data_R_SIR, G1 = simulation_SIR_network(G1, [3],T,beta,mu, False)
plt.plot(data_x_SIR,data_S_SIR,'b',data_x_SIR,data_I_SIR,'r',data_x_SIR,data_R_SIR,'g')
```
